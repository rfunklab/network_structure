```python
graph_type = 'collaboration'
window = '1_1'
collab_data_loc = '/home/gebhart/projects/rfunklab/data/temporal/aps/{}/collaboration/homology'.format(window)

ppy = pd.read_csv('/home/gebhart/projects/rfunklab/data/temporal/aps/1_1/ppy.csv', header=0)
```

```python
def interval_to_numpy(interval):
    '''Helper function to convert parsed homology interval to numpy array'''
    if len(interval) > 1:
        return np.array([list(ival) if len(ival) > 1 else [ival[0],np.inf] for ival in interval])            
    return np.array([])
```

```python
def plot_diagram(diagram, title='', lims=None):
    '''Plots persistence diagram'''
    fig, ax = plt.subplots()
    if diagram.size > 0:
        d = diagram[~np.isinf(diagram[:,1])]
        if d.shape[0] > 0:
            ax.scatter(d[:,0], d[:,1], s=25, c=d[:,1]**2 - d[:,0], cmap=plt.cm.coolwarm, zorder=10)
            if lims is None:
                lims = [
                    np.min(d[:,0]-1),  # min of both axes
                    np.max(d[:,1]+1),  # max of both axes
                ]

            # now plot both limits against eachother
            ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
            ax.set_xlim(lims)
            ax.set_ylim(lims)
    ax.set_aspect('equal')
    plt.xlabel('Birth')
    plt.ylabel('Death')
    plt.title(title)
```

```python
files = [f for f in os.listdir(collab_data_loc) if os.path.isfile(os.path.join(collab_data_loc, f))]
files.sort()
```

```python
# this code block reads the betti numbers, cell counts, and homology information from the .homology files. 
# Set the `doi` variable to change the dimension whose persistence diagram will be  plotted for each year.
# If homology is trivial in that dimension for a given year, nothing will be plotted. 
# Betti numbers and cell counts are printed out, each index in the printed array corresponds to a dimension (starting at 0). 

doi = 1 # the "dimension of interest" to plot persistence diagrams

npints = []
names = []
bettis = []
cell_counts = []
for file in files:
    yr = file[:4]
    print(yr)
    try:
        collab_intervals = parse_intervals(os.path.join(collab_data_loc, file))
        collab_bettis = parse_betti(os.path.join(collab_data_loc,file))
        collab_cell_counts = parse_cell_counts(os.path.join(collab_data_loc,file))
    except IndexError:
        continue
    names.append(yr)
    print('Betti Numbers: ', collab_bettis)
    print('Cell Counts: ', collab_cell_counts)
    bettis.append(collab_bettis)
    cell_counts.append(collab_cell_counts)
    ints = []
    for i in range(len(collab_intervals)):
        collab_npint = interval_to_numpy(collab_intervals[i])
        ints.append(collab_npint)
        if i == doi:
            if collab_npint.size > 0:
                plot_diagram(collab_npint, title='{} Collaboration Dimension {}'.format(yr, i), lims=((0,50)))
            plt.show()
    npints.append(ints)
    
```

```python
# plot betti numbers for each year
betti_up_to_dimension = 10 # set this as the highest dimension to plot (max is 10).

for d in range(betti_up_to_dimension):
    plt.plot(names, [b[d] for b in bettis])
    plt.xticks(rotation=60)
    plt.ylabel('Betti_{}'.format(d))
    plt.xlabel('Year')
    plt.title('Betti Number Dimension {}'.format(d))
    plt.show()
```

```python
# plot betti numbers for each year
betti_up_to_dimension = 10 # set this as the highest dimension to plot (max is 10).

for d in range(betti_up_to_dimension):
    
    fig, ax1 = plt.subplots()

    color = 'red'
    ax1.set_xlabel('Year')
    ax1.tick_params(axis='x', rotation=60)
    ax1.set_ylabel('Normalized Betti_{}'.format(d), color=color)
    ax1.plot(names, [(bettis[b][d]/ppy.iloc[b]['papers']) if len(bettis[b]) > d else 0 for b in range(len(bettis))], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()

    color = 'blue'
    ax2.set_ylabel('Betti_{}'.format(d), color=color)
    ax2.plot(names, [(bettis[b][d]) if len(bettis[b]) > d else 0 for b in range(len(bettis))], color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
    plt.title('Betti Number Dimension {}'.format(d))
    plt.show()

```

```python
# plot cell counts for each year
cc_up_to_dimension = 11 # set this as the highest dimension to plot (max is 11).

for d in range(cc_up_to_dimension):
    plt.plot(names, [c[d] for c in cell_counts])
    plt.xticks(rotation=60)
    plt.ylabel('Cell Count'.format(d))
    plt.xlabel('Year')
    plt.title('Cell Count Dimension {}'.format(d))
    plt.show()
```

```python
# Access the diagrams as numpy arrays in preparation for the following code block. 
doi = 1 # set this to compute Bottleneck and Wasserstein distances in particular "dimension of interest"

diags = []
for i in range(len(npints)):
    npint = npints[i]
    if len(npint) > doi:
        npdoi = npint[doi]
        diags.append(npdoi)
    else:
        diags.append(np.array([]))
```

```python
# WARNING: This code block takes a lot of time to run (~10-45 min depending on `doi` and `yr_ref`).

# This code block computes Bottleneck and Wasserstein distances between each year's diagram and the 
# diagram corresponding to the year in `yr_ref`. This `yr_ref` variable is an integer indexing into
# the array of years, so, for example, yr_ref = 4 --> 1974 since yr_ref = 0 --> 1970. These distance 
# measures are well-defined over the space of persistence diagrams and can be though of as a measure 
# of how different the topology of two spaces are from each other within (or across, though this is 
# less common) dimensions. 

yr_ref = 1 # set this to use different year as reference year. 

bottlenecks = []
wassersteins = []
heats = []
for i in range(yr_ref,len(diags)):
    print('{}/{}'.format(i,len(diags)))
    if diags[i].size > 0:
        diags[i][diags[i] == np.inf] = 50
        bottlenecks.append(persim.bottleneck(diags[yr_ref], diags[i]))
        wassersteins.append(persim.sliced_wasserstein(diags[yr_ref], diags[i]))
#         heats.append(persim.heat(diags[yr_ref], diags[i]))
    else:
        bottlenecks.append(-1)
        wassersteins.append(-1)
#         heats.append(-1)
```

```python
# plot the bottleneck distances for each year with respect to a base year
plt.plot(names[yr_ref:], bottlenecks)
plt.xticks(rotation=60)
plt.ylabel('Bottleneck Distance')
plt.xlabel('Year')
plt.title('Bottleneck Distance ({} Reference)'.format(names[yr_ref]))
```

```python
# plot the wasserstein distances for each year with respect to a base year
plt.plot(names[yr_ref:], wassersteins)
plt.xticks(rotation=60)
plt.ylabel('Wasserstein Distance')
plt.xlabel('Year')
plt.title('Wasserstein Distance ({} Reference)'.format(names[yr_ref]))
```
